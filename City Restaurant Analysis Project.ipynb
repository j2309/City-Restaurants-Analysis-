```python
import pandas as pd
df = pd.read_excel("C:\\Users\\Lenovo\\Desktop\\Dataset .xlsx")   # Replace with your dataset name
df = df.dropna(subset=['City', 'Restaurant Name', 'Cuisines', 'Aggregate rating'])  # Drop missing key data
df = df.drop_duplicates()  # Remove duplicate rows
print("Cleaned dataset shape:", df.shape)
```

```python
### Aggregate rating distributed and the most common rating range. 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip()

# Create rating bins
bins = [0, 1, 2, 3, 4, 5]
labels = ["0-1", "1-2", "2-3", "3-4", "4-5"]
df['Rating Range'] = pd.cut(df['Aggregate rating'], bins=bins, labels=labels, include_lowest=True)

# Count restaurants in each range
rating_distribution = df['Rating Range'].value_counts().sort_index()

print("Rating Distribution:")
print(rating_distribution)

# Find most common rating range
most_common_range = rating_distribution.idxmax()
print(f"\nMost common rating range: {most_common_range}")

# Plot bar chart
plt.figure(figsize=(7,5))
plt.bar(rating_distribution.index.astype(str), rating_distribution.values)
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
### identify the restaurant chain
```

```python
import pandas as pd

# Clean column names
df.columns = df.columns.str.strip()

# Count restaurant occurrences
restaurant_counts = df['Restaurant Name'].value_counts()

# Filter restaurants that appear more than once (chains)
restaurant_chains = restaurant_counts[restaurant_counts > 1].index

# Get cities for each chain
chains_cities = df[df['Restaurant Name'].isin(restaurant_chains)].groupby('Restaurant Name')['City'].unique()

# Convert to DataFrame for nicer display
chains_cities_df = pd.DataFrame({
    'Restaurant Name': chains_cities.index,
    'Cities': chains_cities.values,
    'Number of Branches': df[df['Restaurant Name'].isin(restaurant_chains)].groupby('Restaurant Name').size().values
})

print(chains_cities_df)
```

```python
import folium
from folium.plugins import MarkerCluster

# Initialize map (centered on mean coordinates)
center_lat = df['Latitude'].mean()
center_lon = df['Longitude'].mean()
chain_map = folium.Map(location=[center_lat, center_lon], zoom_start=5)

# Add MarkerCluster
marker_cluster = MarkerCluster().add_to(chain_map)

# Add markers for each chain branch
for _, row in df[df['Restaurant Name'].isin(restaurant_chains)].iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Restaurant Name']} ({row['City']}) - Rating: {row['Aggregate rating']}"
    ).add_to(marker_cluster)

# Display map
chain_map
```

```python
### Ratings and popularity of restaurants
```

```python
import pandas as pd

# Clean column names
df.columns = df.columns.str.strip()

# Identify restaurant chains (appearing more than once)
restaurant_counts = df['Restaurant Name'].value_counts()
restaurant_chains = restaurant_counts[restaurant_counts > 1].index

# Filter only chain restaurants
chains_df = df[df['Restaurant Name'].isin(restaurant_chains)]
# Group by Restaurant Name
chain_stats = chains_df.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',  # Average rating
    'Votes': 'mean',             # Average votes (popularity)
    'City': 'nunique'            # Number of cities present
}).rename(columns={'City': 'Number of Cities'}).sort_values(by='Aggregate rating', ascending=False)

print(chain_stats.head(10))  # Top 10 highest rated chains


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=chain_stats, 
    x='Votes', 
    y='Aggregate rating', 
    size='Number of Cities',  # Size shows how many cities the chain is in
    hue='Aggregate rating',   # Color shows rating
    palette='viridis', 
    sizes=(50, 500),
    legend='brief'
)

plt.title("Ratings vs Popularity of Restaurant Chains")
plt.xlabel("Average Votes (Popularity)")
plt.ylabel("Average Aggregate Rating")
plt.show()
```
